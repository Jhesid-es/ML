### Verificamos que con las columna Area Number Rooms no tiene mucha relaciÃ³n

```python
X_OLS = df.drop(columns=['Avg. Area Number of Rooms','Address','Price']).values
X_train, X_test, y_train, y_test = train_test_split(X_OLS, df['Price'], test_size=0.2, random_state=1)
```

### Entrenamos con regresion lineal multiple

```python
regressor_1 = LinearRegression()
```

```python
#X_train = np.array(X_train).reshape(-1, 1)
#y_train = np.array(y_train).reshape(-1, 1)
#X_test = np.array(X_test).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1, 1)

X_test.shape
```

```python
regressor_1.fit(X_train, y_train)
```

```python
coefficients = regressor_1.coef_

# Obtener el error estÃ¡ndar de la regresiÃ³n
std_error = np.sqrt(np.mean((regressor_1.predict(X_train) - y_train) ** 2))

# Imprimir los coeficientes y el error estÃ¡ndar
for i, coef in enumerate(coefficients):
    print(f'Coeficiente {i+1}: {coef:.4f}')
print(f'Error estÃ¡ndar: {std_error:.4f}')
```

```python
y_test - regressor_1.predict(X_test)
```

### La desviacion de los resultados son grandes, es porque los datos tienen valores mayores a 100000.

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
y_pred = regressor_1.predict(X_test)
```

```python
# Calcular el error cuadrÃ¡tico medio (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calcular el coeficiente de determinaciÃ³n R^2
r2 = r2_score(y_test, y_pred)
```
